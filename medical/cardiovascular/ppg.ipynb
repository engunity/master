# Photoplethysmogram (PPG) pulses

PPG is a non-invasive optical technique used to detect blood volume changes in the microvascular bed of tissue. It is widely used in devices such as pulse oximeters and smartwatches to monitor heart rate and other vital signs. The PPG waveform typically consists of a pulsatile (AC) component that is synchronous with the heartbeat and a slowly varying (DC) component that corresponds to respiration, sympathetic nervous system activity, and thermoregulation.
