# Path Loss

1. Input Parameters:
    - distance (meters)
    - frequency (Hz)
2. Result:
    - Free Space Path Loss (FSPL) in dB.
3. Plot:
    - Compare the performance between 2.4GHz and 5GHz frequencies.

```python
import numpy as np
import matplotlib.pyplot as plt

def free_space_path_loss(distance_meters, frequency_hz):
    # Speed of light in meters per second
    c = 3 * 10**8
    # FSPL formula
    fspl_db = 20 * np.log10(distance_meters) + 20 * np.log10(frequency_hz) - 20 * np.log10(c / (4 * np.pi))
    return fspl_db

# Distance range
distance = np.linspace(1, 1000, 500)  # Distance from 1 meter to 1000 meters

# Frequencies
frequency_2_4ghz = 2.4 * 10**9  # 2.4 GHz
frequency_5ghz = 5 * 10**9  # 5 GHz

# Calculate FSPL for both frequencies
fspl_2_4ghz = free_space_path_loss(distance, frequency_2_4ghz)
fspl_5ghz = free_space_path_loss(distance, frequency_5ghz)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(distance, fspl_2_4ghz, label='2.4 GHz')
plt.plot(distance, fspl_5ghz, label='5 GHz', color='orange')
plt.xlabel('Distance (meters)')
plt.ylabel('Free-Space Path Loss (dB)')
plt.title('Free-Space Path Loss (FSPL) vs Distance')
plt.legend()
plt.grid(True)
plt.show()
```
