# Shannon Limit (Capacity)

How many bits per second of error free information we can send:

C = Channel Capacity [bits/sec]
B = Bandwidth of Channel [Hz]
S = Average Received Signal Power [Watts]
N = Average Noise Power [Watts]

C/B = log2(1+SNR) 

https://pysdr.org/content/channel_coding.html

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the SNR range (in dB)
SNR_dB = np.arange(-10, 30, 0.1)
# Convert SNR from dB to linear scale
SNR_linear = 10**(SNR_dB / 10)
# Calculate the Shannon limit (capacity)
Capacity = np.log2(1 + SNR_linear)

# Plot the Shannon limit
plt.figure(figsize=(10, 6))
plt.plot(SNR_dB, Capacity, label="Shannon Limit")
plt.xlabel("SNR (dB)")
plt.ylabel("Channel Capacity (bits/s/Hz)")
plt.title("Shannon Limit (Capacity vs. SNR)")
plt.grid(True)
plt.legend()
plt.show()
```

The plot above illustrates the relationship between the Signal-to-Noise Ratio (SNR) and the channel capacity.

* 24dB SNR --> 8 bits/s/Hz. If 1MHz  bandwidth is used, 8Mbps. 

The python code below will generate a table with columbs for SNR (dB) and the corresponding channel capacity (bits/s/Hz).

```python
import pandas as pd
import numpy as np

# Define the SNR range (in dB)
SNR_dB = np.arange(-10, 30, 1)
# Convert SNR from dB to linear scale
SNR_linear = 10**(SNR_dB / 10)
# Calculate the Shannon limit (capacity)
Capacity = np.log2(1 + SNR_linear)

# Create a DataFrame
data = {'SNR (dB)': SNR_dB, 'Channel Capacity (bits/s/Hz)': Capacity}
df = pd.DataFrame(data)

# Display the table
print(df)
```
