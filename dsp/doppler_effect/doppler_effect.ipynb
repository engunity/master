# Doppler Effect


1. Input Parameters:
    - Speed of Sound = 343 m/s
    - Source Frequency (Hz)
    - Observer Velocity (m/s)
    - Source Velocity (m/s)
2. Result:
    - Observed Frequency (Hz)


```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
speed_of_sound = 343  # Speed of sound in air in m/s
source_frequency = 1000  # Frequency of the source in Hz

# Function to calculate observed frequency
def doppler_effect(source_velocity, observer_velocity, source_frequency, speed_of_sound):
    observed_frequency = source_frequency * ((speed_of_sound + observer_velocity) / (speed_of_sound - source_velocity))
    return observed_frequency

# Source and observer velocities (you can adjust these)
source_velocity = np.linspace(-50, 50, 400)  # Source moving towards and away from observer
observer_velocity = 0  # Observer is stationary

# Calculate observed frequencies
observed_frequencies = doppler_effect(source_velocity, observer_velocity, source_frequency, speed_of_sound)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(source_velocity, observed_frequencies)
plt.title('Doppler Effect: Observed Frequency vs. Source Velocity')
plt.xlabel('Source Velocity (m/s)')
plt.ylabel('Observed Frequency (Hz)')
plt.grid(True)
```
